```python
fileName = "001_Wen"
loc = "/Users/erfanpakdamanian/Desktop/Research/Research/Platoon/Data/ + fileName + ".xlsx"

dt = pd.ExcelFile(loc) 
sheetNames = dt.sheet_names
df = dt.parse(sheetNames[0])
nData = df.loc[4:df.shape[0]].values
nData = pd.DataFrame(data=nData[1:,0:], columns=nData[0,0:])  

print("\nNumber of up/down", nData.MouseEvent.value_counts())

## get the index of down event
upRowsIndex = np.where(nData['MouseEvent']=='WM_LBUTTONUP')[0].tolist()
downRowsIndex = np.where(nData['MouseEvent']=='WM_LBUTTONDOWN')[0].tolist()

eventSource = 'UI|ABMDeconEEG|ABMRawEEG|ABMThirdParyData'
nData['actualEvent'] = 0
colIndx = nData.columns.get_loc('actualEvent')

for i in range(len(downRowsIndex)-1):
    tmp = nData.iloc[upRowsIndex[i]:downRowsIndex[i+1]]
    upValue = False
    for row in tmp['EventSource']:
        if row == eventSource:
            upValue = True
            break
    if upValue:
       nData.iloc[upRowsIndex[i],colIndx] = 1

nData = nData.drop('LiveMarker', 1)
nData = nData.rename(columns = {'actualEvent':'LiveMarker'})

# these columns should be deleted
rmColumns = ["ESUTimestamp","MsgCounter",	"DataBytes",	"MouseEvent",	"X",	
             "Y","ScrollDelta","XButtonId","KeyCode","NavigationEvent","Url",
             "PageTitle","KeyStroke","MarkerText","SceneType","SceneOutput",
             "SceneParent","MediaTime","PostMarker","Annotation",
             "SDKTimeStamp","StudyName", "ExportDate","Name","Age","Gender",
             "StimulusName","SlideType","EventSource"]

nData = nData.drop(rmColumns, 1)

## you should have a csv file in that folder to overwrite the value on it
csvfile = "/Users/erfanpakdamanian/Desktop/Research/Research/Platoon/Data/liveMarkerEvent_"+fileName+".csv"
nData.to_csv(csvfile)


initData = pd.read_csv("/Users/erfanpakdamanian/Desktop/Research/Research/Platoon/Data/liveMarkerEvent_"+ fileName+'.csv')


#deleting last epoch if it less than 100
epLst = np.unique(initData['Epoc'])
lstEpc = epLst[len(epLst)-1]
if len((initData[initData['Epoc'] == lstEpc])) < 100:
    nwData = initData[initData['Epoc'] != lstEpc]
else:
    nwData = initData

# filtring epochs based on its number    
sumLs256 = 0
sumMr256 = 0
nEpLs = 0
nEpMr = 0
for ep in np.unique(nwData['Epoc']):
    if len((nwData[nwData['Epoc'] == ep])) < 256 :
        nEpLs += 1
        sumLs256 += 256 - len((nwData[nwData['Epoc'] == ep]))
    elif len((nwData[nwData['Epoc'] == ep])) > 256 :
        nEpMr += 1
        sumMr256 += len((nwData[nwData['Epoc'] == ep])) - 256

#shrinking or expanding to exact 256 entries per epoch
diffRowNumber = sumMr256 - sumLs256    
nCutRows = 0
tmpEpc = pd.DataFrame()
if diffRowNumber > 0:
    nCutRows = round(diffRowNumber/nEpMr)
    rowCnt = 0
    for ep in np.unique(nwData['Epoc']):
        if len((nwData[nwData['Epoc'] == ep])) > 256:

            idxStart = min(nwData[nwData['Epoc'] == ep].index)
            idxEnd = max(nwData[nwData['Epoc'] == ep].index)

            if rowCnt != diffRowNumber:
                nIdxEnd = idxEnd - nCutRows
                tmpEpc = tmpEpc.append(nwData[initData['Epoc'] == ep].loc[idxStart:nIdxEnd])
                rowCnt += nCutRows 
                if (diffRowNumber - rowCnt) < nCutRows: nCutRows = diffRowNumber - rowCnt
            else:
                tmpEpc = tmpEpc.append(nwData[initData['Epoc'] == ep].loc[idxStart:idxEnd])

        elif len((nwData[nwData['Epoc'] == ep])) <= 256:
            tmpEpc = tmpEpc.append(nwData[nwData['Epoc'] == ep])


csvfile = "/Users/erfanpakdamanian/Desktop/Research/Research/Platoon/Data/Wen-256Epoc.csv"
tmpEpc.to_csv(csvfile) 
```
